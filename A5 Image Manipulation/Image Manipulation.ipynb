```python
import cv2
import numpy as np

cat = cv2.imread('cat.jpg')
dog = cv2.imread('dog.jpg'
)
```

```python
rc, cc = cat.shape[:2]
Mshiftcat = np.float32([[1,0,10], [0,1,10]])
catshiftedx = cv2.warpAffine(cat, Mshiftcat, (cc, rc))
```

```python
rd, cd = dog.shape[:2]
Mshiftdog = np.float32([[1,0,0],[0,1,20]])
dogshiftedy = cv2.warpAffine(dog, Mshiftdog, (cd, rd))

```

```python

center_cat = (cc // 2, rc // 2)
M_rotate_cat = cv2.getRotationMatrix2D(center_cat, -30, 1.0)
cat_rotated = cv2.warpAffine(cat, M_rotate_cat, (cc, rc))
```

```python
center_dog = (cd // 2, rd // 2)
M_rotate_dog = cv2.getRotationMatrix2D(center_dog, 154, 1.0)
dog_rotated = cv2.warpAffine(dog, M_rotate_dog, (cd, rd))
```

```python
newcatheight = rc * 3
catscaledheight = cv2.resize(cat, (cc, newcatheight))
```

```python
new_width_dog = cd // 2
dog_scaled_width = cv2.resize(dog, (new_width_dog, rd))
```

```python
import matplotlib.pyplot as plt
import cv2

def show_side_by_side(title, before, after):

    before_rgb = cv2.cvtColor(before, cv2.COLOR_BGR2RGB)
    after_rgb = cv2.cvtColor(after, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(8,4))
    plt.suptitle(title)

    plt.subplot(1, 2, 1)
    plt.title("Before")
    plt.imshow(before_rgb)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("After")
    plt.imshow(after_rgb)
    plt.axis("off")

    plt.show()

show_side_by_side("a) Cat Shift X", cat, catshiftedx)
show_side_by_side("b) Dog Shift Y", dog, dogshiftedy)
show_side_by_side("c) Cat Rotation -30", cat, cat_rotated)
show_side_by_side("d) Dog Rotation 154", dog, dog_rotated)
show_side_by_side("e) Cat Height x3", cat, catscaledheight)
show_side_by_side("f) Dog Width /2", dog, dog_scaled_width)
```
