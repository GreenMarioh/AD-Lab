```python
import math
import random
```

```python
def load_txt_data(filename, max_rows=3000):
    data = []

    with open(filename, 'r') as file:
        lines = file.readlines()


        for i in range(1, min(len(lines), max_rows + 1)):
            line = lines[i].strip()

            # split by semicolon
            values = line.split(';')

            # IGNORE first column (timestamp)
            numeric_values = values[1:]

            row = [float(v.replace(',', '.')) for v in numeric_values]
            data.append(row)

    return data
```

```python
def train_test_split(X, y, test_ratio=0.2):
    combined = list(zip(X, y))
    random.shuffle(combined)

    split = int(len(combined) * (1 - test_ratio))
    train = combined[:split]
    test = combined[split:]

    X_train, y_train = zip(*train)
    X_test, y_test = zip(*test)

    return list(X_train), list(X_test), list(y_train), list(y_test)
```

```python
def euclidean_distance(a, b):
    dist = 0
    for i in range(len(a)):
        dist += (a[i] - b[i]) ** 2
    return math.sqrt(dist)
```

```python
def knn_predict(X_train, y_train, X_test, k):
    predictions = []

    for test_point in X_test:
        distances = []

        for i in range(len(X_train)):
            d = euclidean_distance(test_point, X_train[i])
            distances.append((d, y_train[i]))

        distances.sort(key=lambda x: x[0])
        neighbors = distances[:k]

        avg = [0] * len(neighbors[0][1])

        for _, target in neighbors:
            for j in range(len(target)):
                avg[j] += target[j]

        avg = [x / k for x in avg]
        predictions.append(avg)

    return predictions
```

```python
def mean_squared_error(y_true, y_pred):
    error = 0
    count = 0

    for i in range(len(y_true)):
        for j in range(len(y_true[i])):
            diff = y_true[i][j] - y_pred[i][j]
            error += diff * diff
            count += 1

    return error / count
```

```python

data = load_txt_data("ElectricityLoad.txt", max_rows=3000)

X, y = create_features(data)

X_train, X_test, y_train, y_test = train_test_split(X, y)

# K value
k = 3

# Predict on a small test batch
X_test_sample = X_test[:20]
y_true = y_test[:20]

y_pred = knn_predict(X_train, y_train, X_test_sample, k)

# Calculate MSE
mse = mean_squared_error(y_true, y_pred)

print("Mean Squared Error:", mse)
```
